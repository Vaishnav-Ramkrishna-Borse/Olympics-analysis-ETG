### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df = pd.read_csv('summer.csv')
data = df.values
lst = []

for Sport in df['Sport'].unique():
    md = 0
    for i in range(len(data)):
        if (data[i][2] == Sport) and (data[i][8] == 'Gold'):
            md += 1
    lst.append([Sport, md])
    
data = pd.DataFrame(lst, columns = ['Sport','Gold'])
data = data.sort_values(by = 'Gold',ascending = False).head()
data.plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (10,5))

print(data)
```

### 3. Which sport is having most number of medals so far? (Top 5)

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df = pd.read_csv('summer.csv')
data = df.values
ls = []
for Year in df['Year'].unique():
    for i in range(len(data)):
        if (data[i][0] == Year) and (data[i][5] == 'IND') and (data[i][8] == 'Gold') and len(ls) == 0:
            ls.append([Year,data[i][1],data[i][2],data[i][3],data[i][4],data[i][5],data[i][6],data[i][7],data[i][8]])
            
print(ls)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df = pd.read_csv('summer.csv')
data = []

for Event in df['Event'].unique():
    data.append([Event, len(df[df['Event'] == Event])])
    
data = pd.DataFrame(data, columns = ['Event','players'])
data = data.sort_values(by = 'players',ascending = False).head()
data.plot(x = 'Event', y = 'players', kind = 'bar', figsize = (10,5))

print(data)
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df = pd.read_csv('summer.csv')
data = df.values
lst = []

for Sport in df['Sport'].unique():
    md = 0
    for i in range(len(data)):
        if (data[i][2] == Sport) and (data[i][6] == 'Women') and (data[i][8] == 'Gold'):
            md += 1
    lst.append([Sport, md])
    
data = pd.DataFrame(lst, columns = ['Sport','Female_Gold_Medals'])
data = data.sort_values(by = 'Female_Gold_Medals',ascending = False).head()
data.plot(x = 'Sport', y = 'Female_Gold_Medals', kind = 'bar', figsize = (10,5))

print(data)
```
